```python
import pandas as pd    
import matplotlib.pyplot as plt

df = pd.read_csv("vendas.csv")
#Definindo o tamanho da figura
plt.figure(figsize=(10, 5))

#Criando um grafico de barras
#Eixo X = Coluna 'Dia'
#Eixo Y = Coluna, 'Vendas' 
plt.bar(df['Dia'], df['Vendas'], color='green')
#Colocando titulos
plt.title('Vendas por dia da semana')
plt.xlabel('Dia')
plt.ylabel('Total de vendas')
#O comando final que "revela" o desenho
plt.show()
```
